```python
# Import bibliotek
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
# 1. Proszę wyświetlić dane i zapoznać się z posczególnymi informacjami.
iris
```

```python
# 2. Wyświetl liczbę wierszy oraz kolumn.
print("Liczba wierszy: {}".format(iris.shape[0]))
print("Liczba kolumn: {}".format(iris.shape[1]))
```

```python
# 3. Wyświetl podstawowe informacje dla poszczególnych kolumn. Skorzystaj z metod describe.
iris.describe()
```

```python
# 4. Zapoznaj się z działaniem metody groupby.
sepal_length_groupby = iris.groupby("sepal length (cm)")
sepal_length_groupby.mean()
```

```python
# 5. Wykorzystując metodę head proszę wyświelić 5 pierwszych wierszy
iris.head(5)
```

```python
# 6. Korzystając z dropna sprawdź, czy baza zawiera brakujące dane
iris.dropna()
```

Wnioski: baza nie zawiera brakujących danych, liczba wierszy i kolumn się nie zmieniła.

```python
# 7. Korzystając z metody sort_values posortuj dane rosnąco względem drugiej kolumny.
iris.sort_values(by=["sepal width (cm)"])
```

```python
# 8. Wyznacz minimalną i maksymalną długość płatka (kolumna 3 - petal length). Podaj indeks tych wartości (min, idxmin).
petal_length = iris.iloc[:, 2]
print("Maksymalna długość płatka = {}, index = {}".format(petal_length.max(), petal_length.idxmax()))
print("Minimalna długość płatka = {}, index = {}".format(petal_length.min(), petal_length.idxmin()))
```

```python
# 9. Oblicz odchylenie standardowe dla każdej kolumny (std)
iris.std()
```

```python
# 10. Wyodrębnij wiersze dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości (cały zbiór)
sepal_length_mean = iris["sepal length (cm)"].mean()
selected_rows =  iris.loc[iris["sepal length (cm)"] > sepal_length_mean]
selected_rows
```

```python
# 11. Wyświetl histogramy dla poszczególnych parametrów z uwzglednieniem przynależności do danej klasy
iris_target_0 = iris.loc[iris["target"] == 0]
iris_target_1 = iris.loc[iris["target"] == 1]
iris_target_2 = iris.loc[iris["target"] == 2]
ax1 = iris_target_0.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(12,8))
plt.suptitle("Setosa")
```

```python
ax1 = iris_target_1.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(12,8))
plt.suptitle("Versicolor")
```

```python
ax2 = iris_target_2.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(12,8))
plt.suptitle("Virginica")
```
