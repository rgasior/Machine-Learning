Import pakietów

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import danych (baza danych Iris)

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()

#konwersja danych do pakietu Panda
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

### 8. Wyznacz minimalną i maksymalną długość płatka (kolumna 3 - petal length). Podaj indeks tych wartości (min, idxmin).

```python
y = iris.loc[:,"petal length (cm)"]

minimum = y.min()
indeks1 = y.idxmin()
print("Minimalna długość płatka:", minimum, ", indeks wartości:", indeks1, "\n")

maximum = y.max()
indeks2 = y.idxmax()
print("Maksymalna długość płatka:", maximum, ", indeks wartości:", indeks2)
```

### 10. Wyodrębnij wiersze dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości    (cały zbiór)

```python
print("Średnia długość:", iris['sepal length (cm)'].mean())
iris.loc[iris['sepal length (cm)'] > iris['sepal length (cm)'].mean()]
```

### 11. Wyświetl histogramy dla poszczególnych parametrów z uwzglednieniem przynależności do danej klasy

```python
setosa = iris.loc[iris["target"] == 0]
versicolor = iris.loc[iris["target"] == 1]
virginica = iris.loc[iris["target"] == 2]
hist_1 = setosa.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(16,8))
plt.suptitle("Klasa Setosa")
```

```python
hist_2 = versicolor.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(16,8))
plt.suptitle("Klasa Versicolor")
```

```python
hist_3 = virginica.hist(column=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"], figsize=(16,8))
plt.suptitle("Klasa Virginica")
```
