```python
import pandas as pd
import numpy as np
import scipy
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

#Task 1

from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns = iris['feature_names'])
y = pd.DataFrame(iris.target)

data
```

```python
y
```

```python
# 1. Standaryzacja danych (odjęcie średniej, podzielenie przez odchylenie standardowe - względem cechy).
from sklearn.preprocessing import scale

data.replace([np.inf, -np.inf], np.nan)
data.dropna()

data = pd.DataFrame(scale(data), index=data.index, columns=data.columns)
# for i in data:
#     data[i] = (data[i] - data[i].mean()) / data[i].std()

print("\n", "mean:\n", data.mean(), "\n")
print("var:\n", data.var())
data
```

```python
# 2. Obliczenie macierzy kowariancji (numpy.cov() albo X.T@X/m, gdzie X to macierz z danymi postaci mxn: m - ilość przykładów; n - ilość cech.
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
covariance_matrix = np.cov(data.T)
print("covariance_matrix:\n", covariance_matrix)
print("covariance_matrix.shape", covariance_matrix.shape)
```

```python
# 3. Obliczenie wartości i wektorów własnych 
from scipy.linalg import eig
eigenvalues, eigenvectors = eig(covariance_matrix)
print("eigenvalues:\n", eigenvalues)


print("eigenvectors:\n", eigenvectors)
print(eigenvectors.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Wczytanie zbioru danych Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Standaryzacja danych
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_standardized = (X - X_mean) / X_std
```

```python
X_mean
```

```python
X_standardized
```

```python
# Obliczenie macierzy kowariancji
cov_matrix = np.cov(X_standardized.T)
cov_matrix
```

```python
# Obliczenie wartości i wektorów własnych
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
# Wybór dwóch najważniejszych wektorów własnych
top_eigenvectors = eigenvectors[:, :2]

# Redukcja wymiarowości
X_reduced = X_standardized.dot(top_eigenvectors)

# Wykres 2D
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Iris Dataset (2D)')
plt.colorbar()
plt.show()

# Wybór trzech najważniejszych wektorów własnych
top_eigenvectors = eigenvectors[:, :3]

# Redukcja wymiarowości
X_reduced = X_standardized.dot(top_eigenvectors)

# Wykres 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA - Iris Dataset (3D)')
plt.show()
```

```python
# Wybór jednego najważniejszego wektora własnego
top_eigenvector = eigenvectors[:, 0]

# Redukcja wymiarowości
X_reduced = X_standardized.dot(top_eigenvector)

# Wykres 1D
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced, np.zeros_like(X_reduced), c=y)
plt.xlabel('Principal Component 1')
plt.title('PCA - Iris Dataset (1D)')
plt.yticks([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Wczytanie danych
iris = load_iris()
X = iris.data

# Standaryzacja danych
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Obliczenie macierzy kowariancji
X_cov = np.cov(X_scaled.T)

# Obliczenie wartości i wektorów własnych
eigenvalues, eigenvectors = np.linalg.eig(X_cov)

# Sortowanie wartości własnych w malejącej kolejności
eigen_pairs = [(np.abs(eigenvalues[i]), eigenvectors[:, i]) for i in range(len(eigenvalues))]
eigen_pairs.sort(key=lambda x: x[0], reverse=True)

# Wybór k cech z największymi wartościami własnymi
k = 3
w = np.hstack([eigen_pairs[i][1].reshape(4, 1) for i in range(k)])

# Redukcja wymiarowości
Z = X_scaled.dot(w)

# Wykresy
colors = ['navy', 'turquoise', 'darkorange']
target_names = iris.target_names

# Przekształcenie i redukcja wymiarowości dla 1D
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(Z[iris.target == i, 0], np.zeros_like(Z[iris.target == i, 0]), color=color, label=target_name)
plt.title('PCA - Redukcja wymiarowości do 1D')
plt.legend()
plt.show()

# Przekształcenie i redukcja wymiarowości dla 2D
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(Z[iris.target == i, 0], Z[iris.target == i, 1], color=color, label=target_name)
plt.title('PCA - Redukcja wymiarowości do 2D')
plt.legend()
plt.show()

# Przekształcenie i redukcja wymiarowości dla 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax.scatter(Z[iris.target == i, 0], Z[iris.target == i, 1], Z[iris.target == i, 2], color=color, label=target_name)
ax.set_title('PCA - Redukcja wymiarowości do 3D')
ax.legend()
plt.show()

# Porównanie z oryginalnymi cechami obciętymi do wymaganej ilości
X_reduced = X_scaled.dot(w)
X_original_reduced = X[:, :k]

print("Przykładowe przykłady po przekształceniu i redukcji wymiarowości:")
```

```python
# Przekształcenie i redukcja wymiarowości dla 2D (inna perspektywa)
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(Z[iris.target == i, 1], Z[iris.target == i, 2], color=color, label=target_name)
plt.title('PCA - Redukcja wymiarowości do 2D (inna perspektywa)')
plt.legend()
plt.show()

# Przekształcenie i redukcja wymiarowości dla 3D (inna perspektywa)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax.scatter(Z[iris.target == i, 0], Z[iris.target == i, 2], Z[iris.target == i, 1], color=color, label=target_name)
ax.set_title('PCA - Redukcja wymiarowości do 3D (inna perspektywa)')
ax.legend()
plt.show()
```
